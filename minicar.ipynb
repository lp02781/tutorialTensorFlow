import library from tensorflow, pandas, numpy, matplotlib

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as prediction
```

```python
df = pd.read_csv('~/tutorialTensorFlow/model.csv')
df.shape
```

```python
data = df.values
X_data = data[:,:5]
Y_data = data[:,5:]

X = X_data[0:700]
Y = Y_data[0:700]
X_test = X_data[700:873]
Y_test = Y_data[700:873]
```

```python
x_label = np.arange(0,700,1)
plt.figure()
plt.subplot(511)
plt.plot(x_label, X[:,0], '-b')
plt.subplot(512)
plt.plot(x_label, X[:,1], '-b')
plt.subplot(513)
plt.plot(x_label, X[:,2], '-b')
plt.subplot(514)
plt.plot(x_label, X[:,3], '-b')
plt.subplot(515)
plt.plot(x_label, X[:,4], '-b')

plt.show()
```

```python
x_label = np.arange(0,700,1)
plt.figure()
plt.subplot(511)
plt.plot(x_label, Y[:,0], '-r')
plt.subplot(512)
plt.plot(x_label, Y[:,1], '-r')
plt.subplot(513)
plt.plot(x_label, Y[:,2], '-r')
plt.subplot(514)
plt.plot(x_label, Y[:,3], '-r')
plt.subplot(515)
plt.plot(x_label, Y[:,4], '-r')

plt.show()
```

```python
model = tf.keras.models.Sequential([
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      #tf.keras.layers.Dropout(0.2),
      tf.keras.layers.Dense(5, activation= tf.keras.activations.linear)
    ])
model.compile(optimizer= tf.keras.optimizers.Adam(learning_rate =0.001),loss='mean_squared_error',metrics=['accuracy'])
```

```python
learn = model.fit(X, Y, epochs= 5000)
```

```python
model.evaluate(X_test,  Y_test, verbose=2)
```

```python
plt.plot(learn.history['loss'],label = 'Training')
plt.legend()
plt.show()
```

predict output using model

```python
Y_pred = model.predict(X_test)
```

Plot predict result

```python
x_label = np.arange(0,173,1)
plt.figure()
plt.subplot(511)
plt.plot(x_label, Y_test[:,0], '-b', x_label, Y_pred[:,0], '-r')
plt.subplot(512)
plt.plot(x_label, Y_test[:,1], '-b', x_label, Y_pred[:,1], '-r')
plt.subplot(513)
plt.plot(x_label, Y_test[:,2], '-b', x_label, Y_pred[:,2], '-r')
plt.subplot(514)
plt.plot(x_label, Y_test[:,3], '-b', x_label, Y_pred[:,3], '-r')
plt.subplot(515)
plt.plot(x_label, Y_test[:,4], '-b', x_label, Y_pred[:,4], '-r')

plt.show()
```

made a new model

```python
model_new = tf.keras.models.Sequential([
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      tf.keras.layers.Dense(128, activation=tf.nn.relu),
      #tf.keras.layers.Dropout(0.2),
      tf.keras.layers.Dense(2, activation= tf.keras.activations.linear)
    ])
model_new.compile(optimizer= tf.keras.optimizers.Adam(learning_rate =0.001),loss='mean_squared_error',metrics=['accuracy'])
```

load previous model to new model

```python
model_new.load_weights('./model_boat/hehe')
loss,acc = model_new.evaluate(X,  Y, verbose=2)
print("Restored model, accuracy: {:5.2f}%".format(100*acc))
```
